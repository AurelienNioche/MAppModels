```python
import gpytorch
import time
import torch
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import FloatSlider, interact
from scipy.optimize import minimize
from types import SimpleNamespace as Namespace
from scipy.stats import multivariate_normal
```

```python
%config InlineBackend.figure_format='retina'
%matplotlib inline
```

```python
wa = FloatSlider(min=-10, max=1, step=0.01, value=-0.02, description="a")
wb = FloatSlider(min=-1, max=1, step=0.01, value=0.45, description="b")
wc = FloatSlider(min=-1, max=1, step=0.01, value=1.0, description="c")
wd = FloatSlider(min=-1, max=1, step=0.01, value=1.0, description="d")


def f(a=-0.1, b=0.3, c=1.0, d=1.0):
    x =  np.linspace(0, 24, 100)
    y= a*0.01*x**4 + b*x**3 + c*x**2 + d*x
    
    fig, ax = plt.subplots()
    ax.plot(x, y)    

interact(f, a=wa, b=wb, c=wc, d=wd);
```
